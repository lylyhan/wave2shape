```python
from kymatio import Scattering1D
from torch.autograd import backward

learning_rate = 0.1
n_iterations = 100

N = 2**12
scattering = Scattering1D(J=10, shape=(N,))

# Random guess
x = torch.randn((N,), requires_grad=True)
Sx = scattering(x)

# Target
y = torch.randn((N,), requires_grad=False)#was false
Sy = scattering(y)

#with torch.no_grad():
#    x = y.clone()
x.requires_grad = True
print(x.grad)
print(torch.norm(Sx-Sy))

for it_id in range(n_iterations):
    # Backpropagation
    backward(torch.norm(Sx - Sy))
    delta_x = x.grad
    print(x.grad)

    # Gradient descent
    with torch.no_grad():
        x = x - delta_x * learning_rate
    x.requires_grad = True

    # New forward propagation
    Sx = scattering(x)

    # Measure the new loss
    print(torch.norm(Sx - Sy))
```
