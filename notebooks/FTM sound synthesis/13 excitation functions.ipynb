```python
import ftm
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
import time
import ftm_u_shape
import hitdifferentparts
```

```python
w11= 2000#200 * 2 * np.pi #range 200hz-1200hz
tau11 = 1 #range 0.01-0.3
p = 0.3 #how round the sound is, smaller the rougher(metal), range 0-0.3
D = 0 #inharmonicity in smaller values, range 0-10
alpha = 0.1 #range 0-5
sr = 22050
```

```python
def approxnorm(l,mu,s,tau): #normal distribution simulating fx 
    h = l/tau
    #x = np.zeros((1,tau + 1))
    x = []
    for i in range(tau+1):
        #x[i] = 1/(s * np.sqrt(2*np.pi)) * np.exp(-0.5 * (i * h - mu)**2/s**2)
        x.append(1/(s * np.sqrt(2*np.pi)) * np.exp(-0.5 * (i * h - mu)**2/s**2))
    return x
```

```python
x = approxnorm(1,0.5,0.3,300)
l=1
tau = 300
plt.plot(x)
h = l/tau
x2 = np.arange(0,tau+1,1)*h
x2 = np.minimum(2-2/l*x2,2*x2/l)
plt.plot(x2)
plt.show()
```
